```python
import csv
```

```python
f = open('melanoma.csv', 'rb')
reader = csv.reader(f)
year = []
incidence = []

for row in reader: 
    year.append(row[0])
    incidence.append(row[1])
    
    
print year
print incidence
    
f.close()
```

```python
y=year[1:]
print y

a=incidence[1:]
print a

z=map(int,y)
print z

s=map(float,a)
print s
```

```python
from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('the incidence of melanoma per year')
ax.set_xlabel('year')
ax.set_ylabel('incidence')
ax.plot(z,s)
```

the plot showes us that the incidence of melanoma is growing per year
